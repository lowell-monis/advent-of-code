```python
def solution01(input_text):
    def parse_input(input_text):
        rules, updates = input_text.strip().split('\n\n')
        
        rules = {tuple(map(int, rule.split('|'))) for rule in rules.splitlines()}
        
        # Parse updates into lists of integers
        updates = [list(map(int, update.split(','))) for update in updates.splitlines()]
        
        return rules, updates
    
    def is_valid_order(pages, rules):
        # Check each rule that applies to the current update
        for before, after in rules:
            # Skip rules that don't apply to these pages
            if before not in pages or after not in pages:
                continue
                
            # Find positions of the pages
            before_idx = pages.index(before)
            after_idx = pages.index(after)
            
            # If the 'before' page comes after the 'after' page, order is invalid
            if before_idx > after_idx:
                return False
                
        return True
    
    def get_middle_page(pages):
        return pages[len(pages) // 2]
    
    def solve(input_text):
        rules, updates = parse_input(input_text)
        
        # Get middle pages of valid updates
        middle_pages = []
        for update in updates:
            if is_valid_order(update, rules):
                middle_pages.append(get_middle_page(update))
        
        return sum(middle_pages)
    return solve(input_text)
```
